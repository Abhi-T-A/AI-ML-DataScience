```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Cars.csv',skipinitialspace=True)
df
```

```python
df.info()
```

create scatter plot for the following data,make sure all the plots appear in the same plane with x-labels and y-labels evenly spaced
(mpg-hp,mpg-weight)and (hp-mpg,hp-time-to-60)with respect to brand.

```python
fig=plt.subplots(2,2,figsize=(20,10))
plt.subplot(2,2,1) #plot-1
sns.scatterplot(x=df['mpg'],y=df['hp'],data=df,hue=df['brand'])
plt.xlabel('MPG')
plt.ylabel('HP')
plt.title('MPG VS WEIGHT')

plt.subplot(2,2,2) #plot-2
sns.scatterplot(x=df['mpg'],y=df['weightlbs'],data=df,hue=df['brand'])
plt.xlabel('MPG')
plt.ylabel('WEIGHT')
plt.title('MPG VS WEIGHT')

plt.subplot(2,2,3) #plot-3
sns.scatterplot(x=df['hp'],y=df['mpg'],data=df,hue=df['brand'])
plt.xlabel('HP')
plt.ylabel('MPG')
plt.title('HP VS MPG')

plt.subplot(2,2,4) #plot-4
sns.scatterplot(x=df['hp'],y=df['time-to-60'],data=df,hue=df['brand'])
plt.xlabel('HP')
plt.ylabel('TIME_TO_60')
plt.title('HP VS TIME-TO-60')
```

createa bar plots for the following ,make sure all the plots appear in the same plane with xlable and y label evenly spaced.
hp,mpg,weight,time-to-60 with respect to brand and to the number of

hp vs cylinder

mpg vs cylinders

weight and cylinders

time-to-60 vs cylinders


```python
fig=plt.subplots(2,2,figsize=(10,10))
plt.subplot(2,2,1) #plot-1
sns.barplot(x=df['cylinders'],y=df['hp'],data=df,hue=df['brand'])
plt.xlabel('cylinders')
plt.ylabel('HP')
plt.title('cylinders vs hp')

plt.subplot(2,2,2) #plot-2
sns.barplot(x=df['cylinders'],y=df['mpg'],data=df,hue=df['brand'])
plt.xlabel('cylinders')
plt.ylabel('mpg')
plt.title('cylinders vs mpg')

plt.subplot(2,2,3) #plot-3
sns.barplot(x=df['cylinders'],y=df['weightlbs'],data=df,hue=df['brand'])
plt.xlabel('cylinders')
plt.ylabel('weight')
plt.title('cylindes VS weight')

plt.subplot(2,2,4) #plot-4
sns.barplot(x=df['cylinders'],y=df['time-to-60'],data=df,hue=df['brand'])
plt.xlabel('cylinders')
plt.ylabel('TIME_TO_60')
plt.title('HP VS TIME-TO-60')
```

create a box-plot for the following with respect to brand
hp vs cylinder

mpg vs cylinders

weight and cylinders

```python
fig=plt.subplots(2,2,figsize=(20,10))
plt.subplot(2,2,1) #plot-1
sns.boxplot(x=df['cylinders'],y=df['hp'],data=df,hue=df['brand'])
plt.xlabel('cylinders')
plt.ylabel('HP')
plt.title('cylinders vs hp')

plt.subplot(2,2,2) #plot-2
sns.boxplot(x=df['cylinders'],y=df['mpg'],data=df,hue=df['brand'])
plt.xlabel('cylinders')
plt.ylabel('mpg')
plt.title('cylinders vs mpg')

plt.subplot(2,2,3) #plot-3
sns.boxplot(x=df['cylinders'],y=df['weightlbs'],data=df,hue=df['brand'])
plt.xlabel('cylinders')
plt.ylabel('weight')
plt.title('cylindes VS weight')

plt.subplot(2,2,4) #plot-4
sns.boxplot(x=df['cylinders'],y=df['time-to-60'],data=df,hue=df['brand'])
plt.xlabel('cylinders')
plt.ylabel('TIME_TO_60')
plt.title('HP VS TIME-TO-60')
```
