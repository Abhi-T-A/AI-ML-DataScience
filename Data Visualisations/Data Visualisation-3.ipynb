```python
import pandas as pd 
import matplotlib.pyplot as plt

df1 = pd.read_csv('Auto.csv')
df1.head(2)
```

#### Scatter plot

```python
# continuous variables 
# this is ideally used to understand relation between two continuous variables
```

```python
# mileage vs weight 
# these two variables are inversely proportional 
```

```python
plt.scatter(df1['mpg'], df1['weight'])
plt.show()
```

```python
plt.scatter(df1['displacement'], df1['weight'])
```

```python
plt.scatter(df1['acceleration'], df1['displacement'])
plt.show()
```

```python
# reading scatter plot to understand relation will be difficult sometimes 
# we are calculating that relation using correlation
```

```python
df1[['mpg','displacement','acceleration','weight']].corr() # correlation 
# correlation gives math representation for relation between columns
```

## Seaborn

```python
import seaborn as sns
```

```python
# advanced data visualisation 
# you have much control on colors and creating graphs are also easier
```

```python
sns.heatmap(df1[['mpg','displacement','acceleration','weight']].corr() , cmap = 'Blues', annot = True)
```

```python
sns.countplot(x = 'origin', data = df1)
```

```python
sns.countplot(x = 'cylinders', data = df1, hue = 'origin') # hue : color/shade 
```

```python
sns.pairplot(df1[['mpg','acceleration','weight']])
```
