```python
import matplotlib.pyplot as plt
```

```python
A = [23,22,25,23,22]
B = [21,22,23,23,25]
C = [1,2,3,4,5]
```

```python
plt.scatter(C,A)
```

```python
plt.scatter(C,A, color = 'Red', marker = 'D', label = 'Person A') # marker is for shape of dot
plt.scatter(C,B, color = 'Blue', marker = 'o', label = 'Person B' , s = 100) # s is for size parameter
plt.xlabel('Months')
plt.ylabel('Products sold')
plt.title('Sales Performance')
plt.legend() # it will display values written in label pareme
plt.grid() # this enables background lines
plt.show() # to hide this <matplotlib.collections.PathCollection at 0x16221a870>
# always keep plt.show as last line of code
```

```python
plt.subplot(121) # 1 row 2 columns 1st graph
plt.scatter(C,A)
plt.ylabel('products')

plt.subplot(122) # 1 row 2 columns 2nd graph
plt.scatter(C,B)
plt.xlabel('Months')

plt.show()
```

```python
plt.figure(figsize = (10,6)) # width = 10 , height = 6

plt.subplot(221) # 2rows 2 columns 1st graph
plt.scatter(C,A)

plt.subplot(222) # 2rows 2 columns 1st graph
plt.plot(C,A) # line plot

plt.subplot(223) # 2rows 2 columns 1st graph
plt.scatter(C,B)

plt.subplot(224) # 2 rows 2 columns 4th graph
plt.stackplot(C,B)

plt.subplots_adjust(wspace=0.8,hspace = 0.8)
plt.show()
```
