Q1. Which state is having the highest honey production through out all these years.(represent it using a bar graph)?

```python
highest_honey=df.groupby('state')['production'].sum().sort_values(ascending=False).reset_index()
```

```python
highest_honey
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=highest_honey,x='production',y='state',palette='hsv')
plt.show()
```

North Dakota is the highest honey production state.

Q2. Which state is having the highest and lowest yield_per_colony?(represent it using bar graph).

```python
yield_high_low=df.groupby('state')['yield_per_colony'].sum().sort_values(ascending=False).reset_index()
```

```python
yield_high_low
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=yield_high_low,x='yield_per_colony',y='state',palette='hsv')
plt.show()
```

Hawaii is the highest and Maryland is the lowest yield per colony.

Q3. Which year witnesses the highest profit in honey Production?(represent it using bar graph).

```python
profit_honey=df.groupby('year')['value_of_production'].sum().sort_values(ascending=False).reset_index()
```

```python
profit_honey
```

```python
sns.barplot(data=profit_honey,x='year',y='value_of_production',palette='hsv')
plt.xticks(rotation=90)
plt.show()
```

2014

Q5. Which state is having the highest average price in the 21 st century?

```python
df1=df[df['year']>=2020]
avg_price=df1.groupby('state')['average_price'].mean().sort_values(ascending=False).reset_index()
```

```python
avg_price
```

```python
sns.barplot(data=avg_price,x='state',y='average_price',palette='hsv')
plt.xticks(rotation=90)
plt.show()
```

Virginia
