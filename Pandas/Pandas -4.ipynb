```python
def sqr_num(num):
    return num**2
```

```python
sqr_num(6)
```

```python
# def block --> we can reuse the function multiple times across python code
# lambda --> anonymous/ one time function --> which is not stroed in backend so you cant re use it 
```

```python
df1['acceleration'].apply(lambda num: num**2) # pick value from accelration store in num and perform num**2 and repeat 
# the same operation for all the rows
```
