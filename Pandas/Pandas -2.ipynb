```python
import pandas as pd

df1 = pd.DataFrame({'id': [1,2,3,4,5], 'stat_marks': [94,95,97,99,90]}) # 5 rows
df2 = pd.DataFrame({'id': [6,7,8,9,10], 'stat_marks': [91,89,91,88,94]}) # 5 rows
df2 
```

```python
# concat --> union ( sql )
# stacking two data frames on one antoher
```

```python
df3 = pd.concat([df1,df2]) # output of pd.concat is a dataframe stored in a variable name called df3
df3
```

```python
df3.reset_index() # by default older index will be saved a new column #drop = False ( by default )
```

```python
df3.reset_index(drop = True) # drop = True --> older index will not be saved
```

```python
# this is a new concept which doesnt overlap with any other functions from SQL perspective
# it combines data frame using index values
```

```python
df1 = pd.DataFrame({'id': [1,2,3,4,5], 'stat_marks': [94,95,97,99,90]}) # 5 rows
df2 = pd.DataFrame({'id': [3,4,5,6,7], 'ml_marks': [91,89,91,88,94]}) # 5 rows
```

```python
df1
```

```python
df2
```

```python
# join will work based on index 
# get data for 0 index from df1 and combine it with data from df2 where index is 0
```

```python
# suffix : added at end/after
# prefix : added at start/before
```

```python
df1.join(df2, lsuffix = '_df1', rsuffix = '_df2') # when you have common column names in both data frame we need to seperate them by using suffix 
```

```python
# to increase rows : concat 
# to add more columns : merge
```
