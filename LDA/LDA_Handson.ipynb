```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets
```

```python
df=datasets.load_digits()
```

```python
df
```

```python
df.target
```

```python
df.target_names
```

```python
_,axes=plt.subplots(nrows=1,ncols=10,figsize=(15,5))
for ax,image,label in zip(axes,df.images,df.target):
  ax.set_axis_off()
  ax.imshow(image,cmap=plt.cm.grey_r,interpolation='nearest')
  ax.set_title("Training: %i" %label)
```

```python
df.data.shape
```

```python
x=df.data
y=df.target
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Feature scaling
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train.shape
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda=LinearDiscriminantAnalysis(n_components=9)
```

```python
x_train=lda.fit_transform(x_train,y_train)
x_test=lda.transform(x_test)

x_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import *
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
confusion_matrix(y_test,y_pred)
```

```python
cf=confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(cf,annot=True)
```

```python
# deadline 11:40  implement entire file with logistic and decision tree

from sklearn.linear_model import LogisticRegression
```

```python
model_lr=LogisticRegression(random_state=42)
model_lr.fit(x_train,y_train)
y_pred_lr=model_lr.predict(x_test)
accuracy_score(y_test,y_pred_lr)*100
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dt=DecisionTreeClassifier(random_state=42)
model_dt.fit(x_train,y_train)
y_pred_dt=model_dt.predict(x_test)
accuracy_score(y_test,y_pred_dt)*100
```

```python
#Random forest is the best model in terms of accuracy
```
