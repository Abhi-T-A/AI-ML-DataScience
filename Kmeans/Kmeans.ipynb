```python
df_copy=df.copy()
```

```python
df_copy
```

```python
df_copy.drop('Player',axis=1,inplace=True)
```

```python
df_copy
```

```python
# Kmeans --> K --> Elbow curve

from sklearn.cluster import KMeans
```

```python
k_values=[2,3,4,5,6,7,8,9,10]
iner=[]
for k in k_values:
  model=KMeans(n_clusters=k)
  model.fit(df_copy)
  iner.append(model.inertia_)
iner
```

```python
plt.plot(k_values,iner)
```

```python
# k=3
final_model=KMeans(n_clusters=3)
```

```python
final_model.fit(df_copy)
```

```python
final_model.labels_
```

```python
df['ClusterID']=final_model.labels_
```
